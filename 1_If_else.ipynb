### Q3: write a programm to accept a number from 1 to 7 and display the name of the day like 1 for sunday , 2 for monday and so on

```python
number = int(input("enter number:"))
if number<=7:
    if number == 1:
        print("SUNDAY")
    elif number == 2:
        print("MONDAY")
    elif number == 3:
        print("TUESDAY")
    elif number == 4:
        print("WEDNESDAY")
    elif number == 5:
        print("THURSDAY")
    elif number == 6:
        print("FRIDAY")
    elif number == 7:
        print("SATURDAY")
else:
    print("number should be in the range of 1 to 7")
    
```

`Write the if else condition for the given flow:

1. IF AGE IS LESS THAN 30, CHECK IF HE EATS PIZZA OR NOT, IF YES UNFIT ELSE FIT
2. IF AGE IS GREATER THAN 30, CHECK IF HE WILL DO EXCERISE OR NOT, IF YES FIT ELSE UNFIT.

```python
age = int(input("enter your age: "))
if age<30:
    pizza = input("do you eat pizza: ")
    pizza = pizza.lower()
    if pizza == "yes":
        print("UNFIT")
    elif pizza == "no":
        print("FIT")
else:
    execise = input("do you execise: ")
    execise = execise.lower()
    if execise == "yes":
        print("FIT")
    elif execise == "no":
        print("UNFIT")
```

#### PRIME OR NOT

```python
n = int(input("enter number: "))
if n < 2:
    print("not prime")
else:
    for i in range(2, n):
        if n%i == 0:
            print(" not prime")
            break
    else:
        print("prime")
```
